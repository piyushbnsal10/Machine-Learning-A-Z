```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('mediacompany.csv')
data=data.drop('Unnamed: 7',axis =1)
data.head()
```

```python
data['Date']=pd.to_datetime(data['Date'])
data.head()
```

```python
from datetime import date

ref_date=date(2017,2,28)
cur_date=data.Date
date_change=cur_date-ref_date
data['day']=date_change

data.head()
```

```python
data['day']=data['day'].astype(str)
data['day']=data['day'].map(lambda x:x[:2])
data['day']=data['day'].astype(int)
data.head()
```

```python
data.plot.line(x='day',y='Views_show',color='r',label='games')
```

```python
data.plot.line(x='day',y='Views_show',color='b',label='View_show')
data.plot.line(x='day',y='Ad_impression',color='r',label='Ad_impression')
```

```python
data['weekday']=(data['day']+3)%7
data['weekday']=data['weekday'].replace(0,7)
data.head()
```

```python
#MODEL 1 ----------> WEEKDAY And VISITORS

from sklearn.linear_model import LinearRegression
x=data[['Visitors','weekday']]
y=data['Views_show']

LinReg=LinearRegression()
LinReg.fit(x,y)
```

```python
import statsmodels.api as sm
#Unlike SKLearn, statsmodels don't automatically fit a constant, 
#so you need to use the method sm.add_constant(X) in order to add a constant. 
X = sm.add_constant(x)
# create a fitted model in one line
lm_10 = sm.OLS(y,x).fit()
print(lm_10.summary())
```

```python
def cond(i):
    if i % 7 == 5: return 1
    elif i % 7 == 4: return 1
    else :return 0
    return i

data['weekend']=[cond(i) for i in data['day']]
data.head(20)
```

```python
# Putting feature variable to X
data['Ad_impression']=data['Ad_impression']/1000000
x = data[['weekend','Character_A','Ad_impression']]

# Putting response variable to y
y = data['Views_show']

LinReg=LinearRegression()
LinReg.fit(x,y)
```

```python
import statsmodels.api as sm
#Unlike SKLearn, statsmodels don't automatically fit a constant, 
#so you need to use the method sm.add_constant(X) in order to add a constant. 
#X = sm.add_constant(x)
# create a fitted model in one line
lm_10 = sm.OLS(y,x).fit()
print(lm_10.summary())
```

```python
X = data[['weekend','Character_A','Ad_impression']]
Predicted_views = lm_10.predict(X)

from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(data.Views_show, Predicted_views)
r_squared = r2_score(data.Views_show, Predicted_views)

print('Mean_Squared_Error :' ,mse)
print('r_square_value :',r_squared)
```

```python
#Actual vs Predicted
c = [i for i in range(1,81,1)]
fig = plt.figure()
plt.plot(c,data.Views_show, color="blue", linewidth=2.5, linestyle="-")
plt.plot(c,Predicted_views, color="red",  linewidth=2.5, linestyle="-")
fig.suptitle('Actual and Predicted', fontsize=20)              # Plot heading 
plt.xlabel('Index', fontsize=18)                               # X-label
plt.ylabel('Views', fontsize=16)                               # Y-label
```

```python
# Error terms
c = [i for i in range(1,81,1)]
fig = plt.figure()
plt.plot(c,data.Views_show-Predicted_views, color="blue", linewidth=2.5, linestyle="-")
fig.suptitle('Error Terms', fontsize=20)              # Plot heading 
plt.xlabel('Index', fontsize=18)                      # X-label
plt.ylabel('Views_show-Predicted_views', fontsize=16)                # Y-label
```
